```python
import cv2
from google.colab.patches import cv2_imshow

img1 = cv2.imread('bg04.jpg', 1)
# print(img1.shape) # (876, 1400, 3)

# 就是圖片中每一格的 BGR 數值在運算

# 圖像數值基本運算
# 結果圖像 = cv2.運算 (圖像變數一 圖像變數二)
img2 = cv2.add(img1,(255,255,255,0))  # 相加
img2 = cv2.subtract(img1, img2)    # 相減
img2 = cv2.absdiff(img1, (0,0,255,0)) # 相減 絕對值運算

img2 = cv2.divide(img1, (5,5,1,0)) # 相除 四捨五入
img2 = cv2.multiply(img1, (255,255,255,0)) # 相乘

# 用 255 去減數值
img2 = cv2.bitwise_not(img1) # 二進位 not 運算

# 縮放 - 指定大小
img2 = cv2.resize(img1, (800,600))
# 縮放 - 等比率
x = 2
h = int(img1.shape[0] / x)
w = int(img1.shape[1] / x)
img2 = cv2.resize(img1, (w,h))

# 圖片翻轉 翻轉參數 1->左右 0->上下 -1->上下左右
img2 = cv2.flip(img1, -1)

# 圖片旋轉
# 變換矩陣 = cv2.getRotationMatrix2D(中心位置, 角度, 縮放比率)
matrix = cv2.getRotationMatrix2D((200,100), 180, 0.5)
# 結果圖像 = cv2.warpAffine(圖像變數, 變換矩陣, 輸出圖像大小) 旋轉後無數值的地方會留黑(0,0,0)
img2 = cv2.warpAffine(img1, matrix, (1400,876))

# 裁切 使用 numpy 矩陣函式庫計算
# 圖像變數 [ Y 軸起始 : Y 軸結束, X軸起始 : X軸結束 ]
# 圖像變數 [ Y 軸起始 : Y 軸結束 : 間隔 , X軸起始 : X軸結束 : 間隔]
# img2[100:200, 100:200] = img1[0:100 ,0:100]
img2[100:100+img1.shape[0]:2, 100:100+img1.shape[1]:2] = img1[::2, ::2]


# 白平衡  取平均的方法 -> 圖像變數.mean() 
Bavg=img1[:,:,0].mean() 
Gavg=img1[:,:,1].mean() # 取出綠色數值做平均
Ravg=img1[:,:,2].mean()

img2 = img1.copy()
# 圖像變數二 = 圖像變數一的 各色 * 各色加權平均
img2[:,:,0] = cv2.multiply(img1[:,:,0], ( (Bavg+Gavg+Ravg)/(Bavg*3) ))ㄞ
img2[:,:,1] = cv2.multiply(img1[:,:,1], ( (Bavg+Gavg+Ravg)/(Gavg*3) ))
img2[:,:,2] = cv2.multiply(img1[:,:,2], ((Bavg+Gavg+Ravg)/(Ravg*3) ))


cv2_imshow(img2)
```

```python
img1 = cv2.imread('bg04.jpg', 1)

# 影像二值化
# 所有數值都會依造門檻值做篩選，篩選結果由選定 方法 決定

# 二值化方法
# cv2.THRESH_BINARY   # 超過門檻值的像素設為最大值，小於的設為 0
# cv2.THRESH_BINARY_INV # 超過門檻值的像素設為 0 ，小於的設為最大值
# cv2.THRESH_OTSU    # 自動計算門檻值來做二值化，可聯集其他方法使用(只接受單一色彩)

# 回傳門檻值 , 結果圖像 = cv2.threshold() 圖像變數 , 門檻值 , 結果最大值 , 方法)
th, img2 = cv2.threshold(img1, 90, 255, cv2.THRESH_BINARY)
print(th) # 傳回門檻值 90

# 用 cv2.THRESH_OTSU 時，門檻值一定要給，但沒有用
img3 = img1.copy()
th, img3[::,0] = cv2.threshold(img1[::,0], 90, 255, cv2.THRESH_OTSU)
print(th) # 傳回 0
th, img3[::,1] = cv2.threshold(img1[::,1], 90, 255, cv2.THRESH_BINARY|cv2.THRESH_OTSU)
print(th)
th, img3[::,2] = cv2.threshold(img1[::,2], 90, 255, cv2.THRESH_OTSU)
print(th)

# cv2_imshow(img2)
# cv2_imshow(img3)

# 將整張圖分成數個小區塊，自動計算門檻值來做二值化(只接受單一色彩)
# 結果圖像 = cv2.adaptiveThreshold(圖像變數, 最大值, 方法一, 方法二, 區塊大小, 微調值)

# 方法一
# cv2.ADAPTIVE_THRESH_MEAN_C   # 計算區塊大小內的平均值再減去微調值
# cv2.ADAPTIVE_THRESH_GAUSSIAN_C # 計算區塊大小內的高斯加權平均值值再減去微調值
# 方法二
# cv2.THRESH_BINARY
# cv2.THRESH_BINARY_INV

# 結果圖像 = cv2.adaptiveThreshold(圖像變數, 最大值, 方法一, 方法二, 區塊大小, 微調值)
img3=img1.copy()
img3[:,:,0] = cv2.adaptiveThreshold(img1[:,:,0], 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 101, 5)
img3[:,:,1] = cv2.adaptiveThreshold(img1[:,:,1], 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 101, 5)
img3[:,:,2] = cv2.adaptiveThreshold(img1[:,:,2], 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 101, 5)

# cv2_imshow(img3)
```

```python
img1 = cv2.imread('bg04.jpg', 1)

# 影像邊緣偵測
# 結果圖像 = cv2.Canny(來源圖像, 最小門檻值, 最大門檻值)
img2 = cv2.Canny(img1, 100, 200) # 通常會先轉灰階做高斯平滑再用 Canny
# Canny 方法會傳回二維灰階數值


# 影像模糊化
# 平均值模糊法
# 結果圖像 = cv2.blur(圖像變數, 範圍大小)
img2 = cv2.blur(img1, (5,5))

# 中值模糊法
# 結果圖像 = cv2.medianBlur(圖像變數, '單數'的處理數量)
img2 = cv2.medianBlur(img1, 5) # 會取數量中間的值，偶數會跳錯誤


# 影像銳利化
# 直方圖均衡化法，只接受單一色彩
# 結果圖像 = cv2.euqalizeHist(來源圖像)
img2[:,:,0] = cv2.equalizeHist(img1[:,:,0])
img2[:,:,1] = cv2.equalizeHist(img1[:,:,1])
img2[:,:,2] = cv2.equalizeHist(img1[:,:,2])

cv2_imshow(img1)
cv2_imshow(img2)
# print(img2)
# cv2_imshow(m3)
```

```python
import numpy as np

img1 = cv2.imread('bg04.jpg', 1)

# 形態學
# 侵蝕 (色彩低的會侵蝕色彩高的)
# 結果圖像 = cv2.erode(來源圖像, 範圍大小的陣列)
img2 = cv2.erode(img1, np.ones((5,5)))
# 膨脹 (色彩高的會侵蝕色彩低的)
# 結果圖像 = cv2.dilate(來源圖像, 範圍大小的陣列)
img2 = cv2.dilate(img1, np.ones((5,5)))

# 常用去雜訊的方法 -> 先侵蝕後膨脹 or 先膨脹後侵蝕
# 結果圖像 = cv2.morophologyEx(來源圖像, 方法, 範圍大小陣列)
img2 = cv2.morphologyEx(img1, cv2.MORPH_OPEN, np.ones((2,1)))
# 方法
# cv2.MORPH_OPEN   # 先侵蝕後膨脹
# cv2.MORPH_CLOSE   # 先膨脹後侵蝕
# cv2.MORPH_GRADIENT # 侵蝕與膨脹後的變化差


# 色彩篩選 -> 回傳一個二維灰階數值 範圍內為白,其餘為黑
# 結果圖像 = cv2.inRange(img1, 數值下陷限, 數值上限)
img2 = cv2.inRange(img1, (0,0,0), (250,250,250)) # 可單一色彩也可三色彩
img2 = cv2.bitwise_not(img2)

cv2_imshow(img2)

# 應用
# m2=cv2.imread("2.png",1)

# # m3=cv2.absdiff(m2,(36,28,237,0))
# m3=cv2.absdiff(m2,(76,177,34,0))
# m3=cv2.multiply(m3,(255,255,255,255))
# m3=cv2.erode(m3, np.ones((3,3)))
# m3=cv2.bitwise_not(m3)
# m3=cv2.add(m2,m3)
```
